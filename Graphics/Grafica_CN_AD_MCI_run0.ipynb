```python
import matplotlib.pyplot as plt
import numpy as np


import csv

epoch = [] 
acc = []
val_acc = []
loss = []
val_loss = []

with open('Training_CN_AD_MCI_run0.csv','r') as csvfile:
    CSVDB = csv.reader(csvfile, delimiter=';')
    FieldNames = next(CSVDB)

    print(FieldNames)
    
    for row in CSVDB:
        print(row)
        
        epoch.append(int(row[0]))
        acc.append(float(row[1]))
        loss.append(float(row[2]))
        val_acc.append(float(row[3]))
        val_loss.append(float(row[4]))


#************************************************************************************

#acc = history.history['acc']
#val_acc = history.history['val_acc']
#loss = history.history['loss']
#val_loss = history.history['val_loss']

epochs = range(len(acc))


plt.plot(epoch, acc, 'o', label='Training acc')
plt.plot(epoch, val_acc, 'r', label='Validation acc')
plt.legend(['Training acc', 'Validation acc'])
plt.title('Training and validation accuracy')
plt.figure()


plt.plot(epoch, loss, 'o', label='Training loss')
plt.plot(epoch, val_loss, 'r', label='Validation loss')
plt.legend(['Training loss', 'Validation loss'])
plt.title('Training and validation loss')

plt.show()
```
